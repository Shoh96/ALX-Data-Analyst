  #                                        Analysis & Insights Report

> ### WeRateDogs is a Twitter account that rates people's dogs with a humorous comment about the dog. These ratings almost always have a denominator of 10. The numerators, though? Almost always greater than 10. 11/10, 12/10, 13/10, etc. " WeRateDogs has over 7.89M followers (on March 2019) and has received international media coverage.

1. The most common dog_type is the pupper followed by the doggo and they both have the highest (717801) and second highest (571723) followers retweets 209337 and 165740 respectively.

2. The most common dog_breed prediction is golden_retriever followed by pembroke and they both have the highest and second highest retweets respectively.

3. The dog names Stephan, Bo and Duddles had the top three highest retweet counts respectively but there is a difference in rating at the level of the favorite counts.

4. The pupper type has the highest count and the floofer and (doggo|pupper) has the lowest count.

5. The average retweet_count for the doggo is the highest (about 5800 retweets), where the average retweet_count for the floofer is the lowest (less than 2,500).

6. Similarly, the average favorite_count for doggo type is highest (about 20000), and the pupper has the lowest average favorite count.

7. From the dog_breed histogram, we see that the golden retriever is the most common breed in the dataset, followed by the labrador retriever. In third place we have the Pembroke and last but not the least, we have the chihuahua.
